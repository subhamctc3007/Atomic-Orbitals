# 3p - Orbital 

```python
a = 0.529
# 1st Bohr's radius

Z = 1
n = 2
# Principle quantum no.

l = 1
# Azimuthal quantum no.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def R(r):
    global a,Z,n,l
    K = 4/(81*np.sqrt(6))
    return K*( (Z/a)**1.5 )*( ( (6*Z*r)/a - (Z*Z*r*r)/(a*a) )*( np.e**(-Z*r/(3*a)) ) )
```

```python
%matplotlib
res =150
r = np.linspace(0,25*a,res)

psi = []
psi2 = []
for i in range(res):
    psi.append(R(r[i]))
    psi2.append(R(r[i])*R(r[i]))
    
fig = plt.figure()
plt.plot(r,psi, color="blue")
plt.plot(r,0*r, color="orange")

plt.xlabel("r(A)")
plt.ylabel("psi(r)")
plt.title("3p orbital")
plt.grid(True)
plt.show()
```

## Radial Probability Density Function

```python
fig2 = plt.figure()
plt.plot(r,0*r, color="orange")
    
plt.plot(r,psi2, color="blue")

total_prob = 0
for i in range(len(psi2)):
    total_prob += psi2[i]
print(total_prob)

plt.xlabel("r(A)")
plt.ylabel("psi^2(r)")
plt.grid(True)
plt.show()
```

## Radial Probabability Distribution Function

```python
fig3 = plt.figure()

pd = []
for i in range(res):
    pd.append(4*np.pi*r[i]*r[i]*psi2[i])
    
plt.plot(r,0*r, color="orange")
plt.plot(r,pd, color="blue")

plt.xlabel("r(A)")
plt.ylabel("[4(pi)r^2]*psi^2(r)")
plt.title("3p-Orbital")
plt.grid(True)
plt.show()
```

```python
def A_3px(theta,phi):
    return 0.5*np.sqrt(3/np.pi)*np.sin(theta)*np.cos(phi)
# Angular Function
```

```python
theta = np.linspace(0,2*np.pi,100)
phi = np.linspace(0,2*np.pi,100)
x_axis = []
y_axis = []
z_axis = []

for i in range(np.size(theta)):
    for j in range(np.size(phi)):
        for k in range(res):
            
            if psi2[k]*A_3px(theta[i],phi[j])*A_3px(theta[i],phi[j]) >= 0.001:
                
                x_axis.append( r[k]*np.sin(theta[i])*np.cos(phi[j]) )
                y_axis.append( r[k]*np.sin(theta[i])*np.sin(phi[j]) )
                z_axis.append( r[k]*np.cos(theta[i]) )            
```

```python
def draw_axis():
    global r
    ax.plot(r, 0*r, 0*r, color="orange")
    ax.plot(0*r, r, 0*r, color="orange")
    ax.plot(0*r, 0*r, r, color="orange")

    ax.plot(-1*r, 0*r, 0*r, color="orange")
    ax.plot(0*r, -1*r, 0*r, color="orange")
    ax.plot(0*r, 0*r, -1*r, color="orange")
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x_axis, y_axis, z_axis, color='red', marker=".", s=0.05, alpha=0.3)

ax.set_xlim3d(-10,10)
ax.set_ylim3d(-5,5)
ax.set_zlim3d(-5,5)
draw_axis()

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.title("3px Orbital")
plt.show()
```
